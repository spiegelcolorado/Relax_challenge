```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
y = conca_relax['adoption_status']
X = conca_relax.drop('adoption_status', axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
GBC = RandomForestClassifier()
GBC.fit(X_train, y_train)
Y_pred = GBC.predict(X_test)
acc_score = GBC.score(X_test, y_test)
acc_score
```

```python
conca_relax.columns
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, Y_pred))
print(classification_report(y_test, Y_pred))
```

We can see that recall and precision are highly effective with the Random Forest Algorithm. the fact that precision was .98 indicates that there were extremely few false positives. Recall at .76 indicates that the algorithm was able to accurately predict 76% of adopters from the data.

'we can see that the later in time that the account was created the smaller the proportion of people that adopted the app. This could potentially be thrown off because the more recent the user started using the app the less time they have had to use it three time in a seven day span. 'last_session_creation_time' was the most important feature in determining if a user was an adopter. This makes sense because the more recently a person is using the app than the more likely they have adopted the app since they are still using it. One could assume the algorithm is also looking at the difference between creation time and last use time. This is relevant because the longer duration a person continued to use the app the more likely they fully adopted the app.

We can also see that  affillation to organizations also played a large role. This can be seen because,  in terms of feature importance we see that the nature of a person's sign up played a large role in if they became adopters. This can be seen in that 'org_id', and 'invited_by_user_id' had among the most feature importance. This means that certain organizations and people invited by certain people (likely administrators from organizations) were much more likely to adopt the app. Likely because of the importance of communication the app plays for the organization. Similarly people who signed up for personal projects were much less likely to adopt. This can be seen in the feature importance of dummy group 'personal_project'which had a much larger feature importance compared to other dummy group. Further the graph below shows people who were doing personal projects were much less likely to adopt than people whose signed up for reasons of different nature.
